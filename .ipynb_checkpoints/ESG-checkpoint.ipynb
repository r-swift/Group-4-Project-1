```python
lowerq = -1.944722
upperq = 18.001204

iqr = upperq-lowerq
iqr
lower_bound = lowerq - (1.5*iqr)
lower_bound
upper_bound = upperq + (1.5*iqr)
upper_bound

eps_outliers = esg_complete[((esg_complete["AVG Annual EPS"])<=lower_bound)|((esg_complete["AVG Annual EPS"])>upper_bound)]


eps_outliers["ticker"]
```

```python
eps_wo_outliers = esg_complete[esg_complete['ticker'].isin(eps_outliers["ticker"]) == False]
eps_wo_outliers
```

```python
eps_series_wo_outliers= pd.DataFrame(eps_wo_outliers["AVG Annual EPS"])

# Create a boxplot for the data
plt.boxplot(eps_series_wo_outliers)

# Add labels to the x-axis
plt.xticks([1], [""])

# Add a title and labels
plt.title('AVG Annual EPS Overall (without outliers)')
plt.ylabel('AVG Annual EPS')

plt.savefig("Images/AVG Annual EPS Overall (without outliers)BOX.png")

# Show the plot
plt.show()
```

```python
# Define a function to create Linear Regression plots
def linear_regression(x_values, y_values,title):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
    correlation = st.pearsonr(x_values,y_values)
    print(f"The correlation is: {round(correlation[0],2)}")
    print(f"The r-squared is: {rvalue**2}")
    print(f"the p-value is : {pvalue}")
    plt.savefig(f"Images/{title}.png")
    plt.show()
```

```python
# Linear regression on Annualized % Change
plt.title("Annualized % Change by ESG Total Score Overall")
plt.xlabel("Total Score")
plt.ylabel("Annualized % Change") 

linear_regression(perc_change_wo_outliers["total_score"], perc_change_wo_outliers["Annualized % Change"],"Annualized % Change by ESG Total Score Overall" )

```

```python
# Linear regression on AVG Annual EPS
plt.title("Average Annual EPS by ESG Total Score Overall")
plt.xlabel("Total Score")
plt.ylabel("AVG Annual EPS") 
linear_regression(eps_wo_outliers["total_score"], eps_wo_outliers["AVG Annual EPS"],"Average Annual EPS by ESG Total Score Overall")
```

```python
#Grouping the EPS means/medians and then creating a bar graph that shows the average EPS by ESG Grade overall
EPS_group_mean = eps_wo_outliers.groupby(['total_grade']).mean(["AVG Annual EPS"])
EPS_group_mean = EPS_group_mean['AVG Annual EPS']
EPS_group_med = eps_wo_outliers.groupby(['total_grade']).median(["AVG Annual EPS"])
EPS_group_med = EPS_group_med['AVG Annual EPS']

n=4
r = np.arange(n)
width=.25
plt.bar(r,EPS_group_mean, color = 'b', 
        width = width, edgecolor = 'black', 
        label='Mean EPS') 
plt.bar(r+width,EPS_group_med, color = 'g', 
        width = width, edgecolor = 'black', 
        label='Median EPS')
plt.title("Average EPS by ESG Grade Overall")
plt.xlabel("ESG Grade")
plt.ylabel("Average EPS")
plt.xticks(r + width/2,['A','B','BB','BBB']) 
plt.legend() 
plt.show()
```

```python
EPS_A = eps_wo_outliers[eps_wo_outliers['total_grade']=='A']['AVG Annual EPS']
EPS_B = eps_wo_outliers[eps_wo_outliers['total_grade']=='B']['AVG Annual EPS']
EPS_BB = eps_wo_outliers[eps_wo_outliers['total_grade']=='BB']['AVG Annual EPS']
EPS_BBB = eps_wo_outliers[eps_wo_outliers['total_grade']=='BBB']['AVG Annual EPS']

stats.f_oneway(EPS_A, EPS_B, EPS_BB, EPS_BBB)
```

```python
#Grouping the Annualized % means/medians and then creating a bar graph that shows the Annualized % Change by ESG Grade Overall
Perc_group_mean = perc_change_wo_outliers.groupby(['total_grade']).mean(["Annualized % Change"])
Perc_group_mean = Perc_group_mean['Annualized % Change']
Perc_group_med = perc_change_wo_outliers.groupby(['total_grade']).median(["Annualized % Change"])
Perc_group_med = Perc_group_med['Annualized % Change']

n=4
r = np.arange(n)
width=.25
plt.bar(r,Perc_group_mean, color = 'teal', 
        width = width, edgecolor = 'black', 
        label='Mean EPS') 
plt.bar(r+width,Perc_group_med, color = 'purple', 
        width = width, edgecolor = 'black', 
        label='Median EPS')
plt.title("Annualized % Change by ESG Grade Overall")
plt.xlabel("ESG Grade")
plt.ylabel("Annualized % Change")
plt.xticks(r + width/2,['A','B','BB','BBB']) 
plt.legend() 
plt.show()
```
